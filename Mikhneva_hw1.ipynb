```python
from matplotlib.gridspec import GridSpec
import matplotlib.pyplot as plt
import numpy as np
import csv
from matplotlib.font_manager import FontProperties
from matplotlib import font_manager
```

```python
import pandas as pd
df = pd.read_csv('Pearson.tsv', delimiter = '\t')
df
```

```python
font_path = 'cd2f1-36d91_sunday.ttf'  # https://ffont.ru/font/sunday
prop = font_manager.FontProperties(fname=font_path)
```

```python
def plotGraph(title: str):
    x = df['Father']
    y = df['Son']
    plt.scatter(x, y, color = 'darkorchid')
    plt.xlabel("Fathers' height", fontproperties=prop)
    plt.ylabel("Sons' height", fontproperties=prop)
    plt.legend(title_fontsize = 9, title = 'Диаграмма рассеяния роста отцов и сыновей')
```

```python
def gistx1Graph(title: str):
    x = df[df.Father == 67.0].Son
    plt.hist(x, bins= 15, color = 'plum', edgecolor='black')
    plt.xlabel("Sons' height", fontproperties=prop)
    plt.title('while father\'s height is 67 inches', fontsize = 10, fontproperties=prop)
```

```python
def gistx2Graph(title: str):
    x = df[df.Father == 65.0].Son
    plt.hist(x, bins= 15, color = 'plum', edgecolor='black')
    plt.xlabel("Sons' height", fontproperties=prop)
    plt.title('while father\'s height is 65 inches', fontsize = 10, fontproperties=prop)
```

```python
def gisty1Graph(title: str):
    x = df[df.Son == 66.0].Father
    plt.hist(x, bins= 15, color = 'plum', edgecolor='black')
    plt.xlabel("Fathers' height", fontproperties=prop)
    plt.title('while son\'s height is 66 inches', fontsize = 10, fontproperties=prop)
```

```python
if __name__ == '__main__':
    plt.figure(figsize=(18, 14))
    grid = GridSpec(5, 5, wspace = 0.5, hspace = 0.5)
    
    plt.subplot(grid[0: 2, 0: 2])
    plotGraph('Graph 1')

    plt.subplot(grid[0, 2])
    gisty1Graph('Graph 2')

    plt.subplot(grid[0, 3])
    gisty2Graph('Graph 3')
    
    plt.subplot(grid[0, 4])
    gisty3Graph('Graph 4')

    plt.subplot(grid[2, 0])
    gistx1Graph('Graph 5')

    plt.subplot(grid[3, 0])
    gistx2Graph('Graph 6')
    
    plt.subplot(grid[4, 0])
    gistx3Graph('Graph 7')

    plt.show()
```
